```python
import pandas as pd
```

Step 1 : Read the first data (Airport Data)

```python
url = 'https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat'
airport_data = pd.read_csv(url, low_memory=False)
airport_data
```

Step 2 : There aren't table header row. Before we add the table reader row, let's check how many the row and column. After that, let's make it based on https://openflights.org/data#airport

```python
#check total row and column
print("Total Row : ", airport_data.shape[0])
print("Total Column : ", airport_data.shape[1])
```

```python
columns = ["Airport ID", "Name", "City", "Country", "IATA", "ICAO", "Latitude", "Longitude", "Altitude", "Timezone", "DST", "Tz Database", "Type", "Source"]
airport_data = pd.read_csv(url, header=None, names=columns)
print(airport_data)
```

Step 3 : Check if there is any null or error in name, city, or country column

```python
print("Column Name : ", airport_data['Name'].isnull().sum())
print("Column City : ", airport_data['City'].isnull().sum())
print("Column Country : ", airport_data['Country'].isnull().sum())
```

Step 4 : Check Duplicate

```python
for col in ['Name', 'City', 'Country']:
    airport_data[col] = airport_data[col].str.strip().str.upper()

print(airport_data[col])
```

```python
#count distinct of country
print(airport_data['Country'].nunique())
print(airport_data['Country'].unique())
```

Step 4 : Show City that empty in the table

```python
airport_data[(airport_data['City'].isna()) | (airport_data['City'] == '')][['Name', 'City', 'Country']]
```

Step 5 : Make a new file in google sheets that already saved as csv and upload through github to fill the empty city

```python
url2 = 'https://raw.githubusercontent.com/raisharhmn/da_airport/refs/heads/main/Filled%20City.csv'
filled_city_col = ["Name", "City", "Country"]
filled_city= pd.read_csv(url2, header=None, names=filled_city_col)

filled_city
```

Step 6 : Fill the city

```python
merged = airport_data.merge(filled_city, on='Name', how='left', suffixes=('', '_filled'))
merged['City'] = merged['City'].fillna(merged['City_filled'])
merged = merged.drop(columns=['City_filled'])
```

Step 7 : Let's make a new table for airport type (military airport, airport, air base, force air station, etc)

```python
import re

def extract_type(name):
    name_upper = name.upper()
    if "MILITARY AIR BASE" in name_upper:
        return "MILITARY AIR BASE"
    elif "AIR FORCE STATION" in name_upper:
        return "AIR FORCE STATION"
    elif "AIRBASE" in name_upper:
        return "AIRBASE"
    elif "AIRFIELD" in name_upper:
        return "AIRFIELD"
    elif "AIR BASE" in name_upper:
        return "AIR BASE"
    elif "AIRPORT" in name_upper:
        return "AIRPORT"
    else:
        return "OTHER"

merged["Airport_Type"] = merged["Name"].apply(extract_type)
merged
```

Step 9 : There are still a lot of improvement in the library, so let's fill it

```python
import pycountry_convert as pc

def country_to_continent(country_name):
    try:
        country_name = country_name.strip().title()
        if country_name == "Usa":
            country_name = "United States"
        elif country_name == "Uk":
            country_name = "United Kingdom"
        elif country_name == "South Korea":
            country_name = "Korea, Republic of"
        elif country_name == "North Korea":
            country_name = "Korea, Democratic People's Republic of"
        elif country_name == "Russia":
            country_name = "Russian Federation"
        elif country_name == "Vietnam":
            country_name = "Viet Nam"
        elif country_name == "Czech Republic":
            country_name = "Czechia"
        elif country_name == "Cote D'Ivoire":
            country_name = "Ivory Coast"
        elif country_name == "Congo (Kinshasa)":
            country_name = "Congo, Democratic Republic of the"
        elif country_name == "Congo (Brazzaville)":
            country_name = "Congo"
        elif country_name == "Burma":
            country_name = "Myanmar"
        elif country_name == "Saint Helena":
            country_name = "Saint Helena, Ascension and Tristan da Cunha"
        elif country_name == "Reunion":
            country_name = "Réunion"
        elif country_name == "Sao Tome And Principe":
            country_name = "Sao Tome and Principe"
        elif country_name == "Saint Pierre And Miquelon":
            country_name = "Saint Pierre and Miquelon"
        elif country_name == "Bosnia And Herzegovina":
            country_name = "Bosnia and Herzegovina"
        elif country_name == "Turks And Caicos Islands":
            country_name = "Turks and Caicos Islands"
        elif country_name == "Wallis And Futuna":
            country_name = "Wallis and Futuna"
        elif country_name == "Antarctica":
            country_name = "Antarctica"
        elif country_name == "West Bank":
            country_name = "Palestine, State of"
        elif country_name == "Midway Islands":
            country_name = "United States Minor Outlying Islands"
        elif country_name == "Antigua And Barbuda":
            country_name = "Antigua and Barbuda"
        elif country_name == "Virgin Islands":
            country_name = "Virgin Islands, U.S."
        elif country_name == "Saint Kitts And Nevis":
            country_name = "Saint Kitts and Nevis"
        elif country_name == "Trinidad And Tobago":
            country_name = "Trinidad and Tobago"
        elif country_name == "Saint Vincent And The Grenadines":
            country_name = "Saint Vincent and the Grenadines"
        country_code = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        return pc.convert_continent_code_to_continent_name(continent_code)
    except:
        return 'Unknown'

merged['Continent'] = merged['Country'].apply(country_to_continent)
merged
```

Step 10 : Owh, there still Unknown continent. Let's make the new file to fill the continent

```python
merged[(merged['Continent'] == 'Unknown')][['Name', 'City', 'Country','Continent']]
```

```python
url3 = 'https://raw.githubusercontent.com/raisharhmn/da_airport/refs/heads/main/Fill%20the%20Continent.csv'
filled_continent= pd.read_csv(url3)
filled_continent
```

```python
merged2 = merged.merge(filled_continent, on='Country', how='left', suffixes=('', '_filled'))
merged2.loc[merged2['Continent'] == 'Unknown', 'Continent'] = merged2['Continent_filled']
merged2 = merged2.drop(columns=['Continent_filled'])
```
